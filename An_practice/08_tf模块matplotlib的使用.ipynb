```python
import numpy as np
import matplotlib.pyplot as plt
x=np.array([1,2,3,4,5,6,7,8])
y=np.array([3,4,5,7,8,6,10,15])
plt.plot(x,y,'red')#折线 1 x坐标 2 y 坐标 3 颜色
plt.plot(x,y,'green',lw=10)#lw表示折线的宽度,而且会覆盖前一个折线图
x=np.array([1,2,3,4,5,6,7,8])
y=np.array([1,13,2,4,12,12,3,4])
plt.bar(x,y,0.5,alpha=1,color='blue')#柱状图 1 x坐标 2 y坐标 3 宽度比 4 透明度 5 颜色
#柱状图可以与折线图一起展示
plt.show()
```
